```python
minimax = []
evaluation_func = []
endgame_func = []
```

```python
for df in df_all:
    m = df[df["Function name"] == "__minimax"]['Avg time Per Call (tottime/ncalls)'].iloc[0]
    eval_func = df[df["Function name"] == "evaluation_func"]['Avg time Per Call (tottime/ncalls)'].iloc[0]
    diags = df[df["Function name"] == "check_diagonals"]['Avg time Per Call (tottime/ncalls)'].iloc[0]
    horizontal = df[df["Function name"] == "check_horizontal"]['Avg time Per Call (tottime/ncalls)'].iloc[0]
    south = df[df["Function name"] == "check_south"]['Avg time Per Call (tottime/ncalls)'].iloc[0]
    end = df[df["Function name"] == "is_endgame"]['Avg time Per Call (tottime/ncalls)'].iloc[0]

    total_endgame = diags + horizontal + south + end

    minimax.append(m)
    evaluation_func.append(eval_func)
    endgame_func.append(total_endgame)
```

```python
minimax
```

```python

fig, ax = plt.subplots()
ax.plot(np.arange(1, 8), minimax)
ax.grid(True)
plt.show()
```
